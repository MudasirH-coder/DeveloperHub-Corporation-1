```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**lOAD DATASET FROM PYTHON SEABORN LIBRARY **

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

**READING DATASET**

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head(10)
```

```python
df.info
```

**UNDERSTANDING THE DATA **

```python
print(df.shape)
```

**STATISTICAL SUMMARY **

```python
df.describe()
```

**CHECK MISSING VALUES**

```python
df.isnull().sum()
```

**COVARIANCE & CORRELATION**:   Covariance and correlation are two fundamental statistical metrics used to evaluate the relationship between two variables



```python
print('covariance:')
df.cov()
```

**CONVERT SEX COLUMN TO INTGER VALUES**:We convert the sex column to integer values because machine learning algorithms and mathematical models can only process numerical data. Algorithms require mathematical inputs (like distances and weights) and cannot directly interpret text strings such as "male" or "female"

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

# Add the target variable from the Iris dataset to the DataFrame
df['iris_target'] = iris.target

# Now, encode the 'iris_target' column
df['iris_target_encoded'] = lb.fit_transform(df['iris_target'])

# Display the first 5 encoded values
df['iris_target_encoded'][:5]
```

ANALYSIS THE DATA VISUALLY NUMBER OF SPECIES

```python
df['iris_target_encoded'].value_counts()
```

Unlike, Iris dataset which has equal number of all the species of flowers(50 each), the penguins dataset has different number of samples for each species.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))
df['iris_target_encoded'].value_counts().plot(kind='bar')
plt.show()
```

**HEATMAP:**

```python
plt.figure(figsize=(7,2))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
